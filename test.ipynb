```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output_set.csv')
normal_runtime = df['Runtime (s)']

plt.figure(figsize=(24, 6))  # Adjust the width and height as needed

# Visualize a column (e.g., 'Runtime (s)') to a graph
plt.plot(df['Operation'], normal_runtime, marker='o', linestyle='-')
plt.xlabel('Operation')
plt.ylabel('Runtime (s)')
plt.title('Runtime per Operation')
plt.grid(True)
plt.show()
```
