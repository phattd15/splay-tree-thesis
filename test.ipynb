```python
df_depth = pd.read_csv('case_depth.csv')
df_normal = pd.read_csv('case_no_depth.csv')
df_set = pd.read_csv('case_set.csv')

X = df_depth['test_size']
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis - 0.2, df_depth['avg_runtime'], 0.2, label = 'Auto Depth Splay') 
plt.bar(X_axis + 0, df_normal['avg_runtime'], 0.2, label = 'Normal Splay')
plt.bar(X_axis + 0.2, df_set['avg_runtime'], 0.2, label = 'STL Set')
  
plt.xticks(X_axis, X) 
plt.xlabel("Test Size") 
plt.ylabel("Average Runtime (s)") 
plt.title("Benchmark in runtime for multiple test size with random access") 
plt.legend() 
plt.show() 
```

```python
df_depth = pd.read_csv('cache_depth.csv')
df_normal = pd.read_csv('cache_no_depth.csv')
df_set = pd.read_csv('cache_set.csv')

X = df_depth['test_size']
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis - 0.2, df_depth['avg_runtime'], 0.2, label = 'Auto Depth Splay') 
plt.bar(X_axis + 0, df_normal['avg_runtime'], 0.2, label = 'Normal Splay')
plt.bar(X_axis + 0.2, df_set['avg_runtime'], 0.2, label = 'STL Set')
  
plt.xticks(X_axis, X) 
plt.xlabel("Cache Size") 
plt.ylabel("Average Runtime (s)") 
plt.title("Benchmark in runtime for multiple test size with cache access") 
plt.legend() 
plt.show() 
```
